```python
import matplotlib.pyplot as plt
import numpy as np

# Criação dos dados para o mesh grid
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))  # Função de exemplo, poderia ser qualquer outra

# Gerar pontos aleatórios dentro do intervalo dos dados X e Y
num_points = 100
x_points = np.random.uniform(-5, 5, num_points)
y_points = np.random.uniform(-5, 5, num_points)

# Plotagem do gráfico de mesh grid com pontos aleatórios
plt.figure(figsize=(8, 6))
plt.contourf(X, Y, Z, cmap='viridis')  # Para um gráfico de contorno preenchido
plt.colorbar()  # Adiciona uma barra de cores para referência
plt.scatter(x_points, y_points, color='red', s=10, alpha=0.5)  # Plotar os pontos aleatórios
plt.title('Gráfico de Mesh Grid com Pontos Aleatórios')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.show()
```
