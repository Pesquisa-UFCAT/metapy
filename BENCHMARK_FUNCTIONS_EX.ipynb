```python
from META_TOOLBOX import META_FUNCTIONS
import numpy as np
import matplotlib.pyplot as plt

def PLOT_BENCHMARK(X,FUNCTION,X_L, X_U):
    N = 100
    X_AUX = np.linspace(X_L, X_U, N)
    Y_AUX = np.linspace(X_L, X_U, N)
    A, B = np.meshgrid(X_AUX, Y_AUX)
    Z = np.zeros((N,N))
    for I in range(N):
        for J in range(N):
            X = [A[I, J], B[I, J]]
            if FUNCTION == "ACKLEY":
                Z[I, J] = META_FUNCTIONS.ACKLEY(X)
            if FUNCTION == "SPHERE":
                Z[I, J] = META_FUNCTIONS.SPHERE(X)
            if FUNCTION == "ROSENBROCK":
                Z[I, J] = META_FUNCTIONS.ROSENBROCK(X)
            if FUNCTION == "RASTRIGIN":
                Z[I, J] = META_FUNCTIONS.RASTRIGIN(X)
            if FUNCTION == "GRIEWANK":
                Z[I, J] = META_FUNCTIONS.GRIEWANK(X)
            if FUNCTION == "ZAKHAROV":
                Z[I, J] = META_FUNCTIONS.ZAKHAROV(X)
            if FUNCTION == "EASOM":
                Z[I, J] = META_FUNCTIONS.EASOM(X)
            if FUNCTION == "MICHALEWICS":
                Z[I, J] = META_FUNCTIONS.MICHALEWICS(X)
    fig = plt.figure()
    plt.title(FUNCTION)
    ax = plt.axes(projection='3d')
    ax.contour3D(A, B, Z, 100, cmap='jet')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('z');
```

```python
X = [0,0,0]
PLOT_BENCHMARK(META_FUNCTIONS.ACKLEY(X),"ACKLEY",-32.768,32.768)
PLOT_BENCHMARK(META_FUNCTIONS.SPHERE(X),"SPHERE",-5.12,5.12)
PLOT_BENCHMARK(META_FUNCTIONS.ROSENBROCK(X),"ROSENBROCK",-2048,2048)
PLOT_BENCHMARK(META_FUNCTIONS.RASTRIGIN(X),"RASTRIGIN",-5.12,5.12)
PLOT_BENCHMARK(META_FUNCTIONS.GRIEWANK(X),"GRIEWANK",-600,600)
PLOT_BENCHMARK(META_FUNCTIONS.ZAKHAROV(X),"ZAKHAROV",-5,10)
PLOT_BENCHMARK(META_FUNCTIONS.EASOM(X),"EASOM",-100,100)
PLOT_BENCHMARK(META_FUNCTIONS.MICHALEWICS(X),"MICHALEWICS",0,np.pi)
```
