# Golden Section notebook

Import library

```python
from META_TOOLBOX import GOLDEN_SECTION_ALGORITHM
```

## Setup

```python
SETUP = {
        'N_ITER': 50,
        'X_L': -1.0,
        'X_U': 2.0,
        'NULL_DIC': None,
        }
```

## OF statement 

```python
def OF_FUNCTION(x, NULL_DIC):
    OF = x**3 + x**2 - 3*x - 3
    return OF
```

## Call algorithm

```python
BEST_REP = GOLDEN_SECTION_ALGORITHM(OF_FUNCTION, SETUP)
```

## Best result

```python
BEST = BEST_REP[0]
DIMENSIONS = list(BEST['X_POSITION'][-1,:])
print(' x*:     ', DIMENSIONS[-1], 
      '\n of(x*): ', BEST['OF'][-1])
```
