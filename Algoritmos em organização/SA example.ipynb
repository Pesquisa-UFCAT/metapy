## 3.0 OF

```python
# SPHERE
def SPHERE(X):
    """
    Sphere benchmark function D-dimension
    """
    DIM = len(X)
    SUM = 0
    for I_COUNT in range(DIM):
        X_I = X[I_COUNT]
        SUM += X_I ** 2
    Y = SUM
    return Y

# OBJ. Function
def OF_FUNCTION(X, NULL_DIC):
    OF = SPHERE(X)
    return OF
```
