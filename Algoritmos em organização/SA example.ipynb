## 3.0 OF

```python
# FUNÇÃO ROSENBROCK
def ROSENBROCK(X):
    """
    Rosenbrock benchmark function D-dimension
    """
    DIM = len(X)
    SUM = 0
    for I_COUNT in range(DIM - 1):
        X_I = X[I_COUNT]
        X_NEXT = X[I_COUNT + 1]
        NEW = 100 * (X_NEXT - X_I ** 2) ** 2 + (X_I - 1) ** 2
        SUM += NEW
    Y = SUM
    return Y

# OBJ. Function
def OF_FUNCTION(X, NULL_DIC):
    OF = ROSENBROCK(X)
    return OF
```
