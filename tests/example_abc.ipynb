```python
# Build an initial population (list of lists)
rng = np.random.default_rng(42)
n_pop = 10
d = 3
x_lower = -5.12
x_upper = 5.12
initial_population = [(rng.random(d) * (x_upper - x_lower) + x_lower).tolist() for _ in range(n_pop)]

# ABC params
params = {
    'limit': 20,
    'scout': True,
    'employed_updates': 1,
    'onlooker_updates': n_pop,  # same as number of onlookers (common choice)
    'seed': 123
}

# Run the algorithm
df_all, df_resume, report = abc_algorithm_01(
    obj=sphere,
    n_gen=10,
    params=params,
    initial_population=initial_population,
    x_lower=x_lower,
    x_upper=x_upper,
    args=None
)

# Inspect results
print("=== Summary (df_resume) ===")
print(df_resume)
print("\n=== Final best/avg/worst (last row) ===")
print(df_resume.tail(1))
print("\n=== Report snippet ===")
print(report[:1000])  # print the start of the report string

# get the best solution at final iteration from df_all
df_final = df_all[df_all['ITER'] == 10].reset_index(drop=True)
best_idx = int(df_final['OF'].idxmin())
print("\n=== Best solution in final iteration (df_final row) ===")
print(df_final.loc[best_idx])
```
