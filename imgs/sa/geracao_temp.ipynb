```python
import pandas as pd
from EASYPLOT_TOOLBOX import *
```

```python
cont = 1
x = []
geom = []
lundy = []
linear = []
loga = []
n = 50
for i in range (n, 0, -1):
    x.append(cont)
    if i == n:
      geom.append(i)
      lundy.append(i)
      linear.append(i)
      loga.append(i)
    else:
      geom.append(geom[-1]*0.9)
      lundy.append(lundy[-1] / (1 + lundy[-1] * 0.9E-3))
      linear.append(linear[-1] - 0.09 * linear[-1])
      loga.append(loga[-1] * np.exp(- 0.09 * cont))
      cont += 1
```

```python
# Data
DF =  pd.DataFrame({'x0': x,
                    'y0': geom,
                    'x1': x,
                    'y1': lundy,
                    'x2': x,
                    'y2': linear,
                    'x3': x,
                    'y3': loga
                   })

# Chart setup
CHART_CONFIG = {
              'NAME': 'figure1w',
              'WIDTH': 15.0,
              'HEIGHT': 7.5,
              'MARKER': ['none', 'none', 'none', 'none'],
              'MARKER SIZE': 3,
              'LINE WIDTH': 4,
              'LINE STYLE': ['--', '--', '--', '--'],
              'X AXIS LABEL': 'iteration',
              'X AXIS SIZE': 10,
              'Y AXIS LABEL': 'Temperature',
              'Y AXIS SIZE': 10,
              'AXISES COLOR': '#000000',
              'LABELS SIZE': 14,
              'LABELS COLOR': '#000000',
              'CHART COLOR': ['#000000','olive','r', 'b'],
              'ON GRID?': True,
              'LEGEND': ['geometric α=0.9', 'lundy α=0.0009', 'linear α=0.09', 'exp α=0.09'],
              'LOC LEGEND': 'upper right',
              'SIZE LEGEND': 8,
              'Y LOG': False,
              'X LOG': False,
              'DPI': 600,
              'EXTENSION': 'svg'
             }

# Data statement
DATA = {'DATASET': DF}

# Call function
LINE_CHART(DATASET = DATA, PLOT_SETUP = CHART_CONFIG)
```
